```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://drive.google.com/uc?export=download&id=1KfIzbQZwilaAL_W2vr5vcVjbVlYLQOb6"

input_df = pd.read_excel(url, index_col = 0, header = None)
input_df.head()
```

```python
input_df.tail()
```

```python
axes = input_df.plot(marker=None, figsize=(20, 20), subplots = True)
```

```python
#240 str

from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

output_notebook()

first = input_df.iloc[-240:, 0]
second = input_df.iloc[-240:, 1]

track = first + second

p = figure(plot_width = 800, plot_height = 400)
# p.line(np.arange(len(first)), first, line_width = 2)
# p.line(np.arange(len(second)), second, line_width = 2, color = '#F00')

a = []
b = []

for i in range(1, 239):
  if track[i] > track[i-1] and track[i] > track[i+1]:
    a.append(track[i])
    b.append(i)


p.line(np.arange(len(track)), track, line_width = 2, color = 'green')
p.circle(b, a, size=5, color='red')


show(p)
```
